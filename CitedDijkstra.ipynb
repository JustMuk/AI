```python
#Initialize auxiliary arrays
import numpy
import matplotlib
maxnum= 5
distancemap=np.ones((max_val,max_val),dtype=int)*np.Infinity
distancemap[0,0]=0
originmap=np.ones((max_val,max_val),dtype=int)*np.nan
visited=np.zeros((max_val,max_val),dtype=bool)
finished = False
x,y=int(0),int(0)
count=0

#Loop Dijkstra until reaching the target cell
while not finished:
  # move to x+1,y
  if x < max_val-1:
    if distancemap[x+1,y]>map[x+1,y]+distancemap[x,y] and not visited[x+1,y]:
      distancemap[x+1,y]=map[x+1,y]+distancemap[x,y]
      originmap[x+1,y]=np.ravel_multi_index([x,y], (max_val,max_val))
  # move to x,y+1
  if y < max_val-1:
    if distancemap[x,y+1]>map[x,y+1]+distancemap[x,y] and not visited[x,y+1]:
      distancemap[x,y+1]=map[x,y+1]+distancemap[x,y]
      originmap[x,y+1]=np.ravel_multi_index([x,y], (max_val,max_val))
  # move to x,y-1
  if y>0:
    if distancemap[x,y-1]>map[x,y-1]+distancemap[x,y] and not visited[x,y-1]:
      distancemap[x,y-1]=map[x,y-1]+distancemap[x,y]
      originmap[x,y-1]=np.ravel_multi_index([x,y], (max_val,max_val))

  visited[x,y]=True
  dismaptemp=distancemap
  dismaptemp[np.where(visited)]=np.Infinity
  
  minpost=np.unravel_index(np.argmin(dismaptemp),np.shape(dismaptemp))
  x,y=minpost[0],minpost[1]
  if x==max_val-1 and y==max_val-1:
    finished=True
  count=count+1

 
mattemp=map.astype(float)
x,y=max_val-1,max_val-1
path=[]
mattemp[int(x),int(y)]=np.nan

while x>0.0 or y>0.0:
  path.append([int(x),int(y)])
  xxyy=np.unravel_index(int(originmap[int(x),int(y)]), (max_val,max_val))
  x,y=xxyy[0],xxyy[1]
  mattemp[int(x),int(y)]=np.nan
path.append([int(x),int(y)])

#Output and visualization of the path
current_cmap = plt.cm.Blues.copy()
current_cmap.set_bad(color='green')
fig, ax = plt.subplots(figsize=(max_val,max_val))
ax.matshow(mattemp,cmap=plt.cm.Blues, vmin=0, vmax=20)
for i in range(max_val):
    for j in range(max_val):
      c = map[j,i]
      ax.text(i, j, str(c), va='center', ha='center')

print('The path length is: '+str(distancemap[max_val-1,max_val-1]))
```
